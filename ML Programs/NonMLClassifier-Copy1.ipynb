```python
import pandas as pd
```

```python
path = '../Allwords.csv'
pd1 = pd.read_csv(path)
pd1.head()
```

```python
pd2 = pd1.set_index('Negative')
pd2 = pd2.drop('Positive', axis=1)
pd3 = pd1.set_index('Positive')
pd3 = pd3.drop('Negative', axis=1)
pd3.head()
```

```python
def sentScore(split_text):
    posCount = 0
    negCount = 0
    posArray = []
    negArray = []
    for x in split_text:
        try:
            posArray.append(pd3.loc[x])
        except KeyError:
            pass
        try:
            negArray.append(pd2.loc[x])
        except KeyError:
            pass
    posCount = len(posArray)
    negCount = len(negArray)
#     print(posArray)
#     print(negArray)
    return((posCount - negCount)/len(split_text)) #, posCount, negCount)
```

```python
path2 = '../Datafiniti_Hotel_Reviews.csv'
```

```python
df2 = pd.read_csv(path2)
```

```python
df2.head()
```

```python
df33 = df2[['reviews.rating', 'reviews.text']]
```

```python
df33.head()
```

```python
rate_list = []
math_list = []
actual_list = []

for x in range(len(df33)):
    splitter = df33.iloc[x]['reviews.text'].split()
#     print(splitter)
    senty = sentScore(splitter)
    if senty > 0:
        senty2 = 1
    elif senty < 0:
        senty2 = 0
    else:
        senty2 = None
    #print(str(senty2) + ' Actual Rating:  ' + str(df33.iloc[x]['reviews.rating']))
    rate_list.append(senty2)
    math_list.append(senty)
    actual_list.append(df33.iloc[x]['reviews.rating'])
```

```python
df44 = pd.DataFrame(data=rate_list)
df45 = pd.DataFrame(data=math_list)
df46 = pd.DataFrame(data=actual_list)
```

```python
df77 = df44.join(df45, lsuffix='a')
```

```python
df77 = df77.join(df46, rsuffix='b')
```

```python
df77.set_axis(['Guess', 'Score', 'Rating'], axis=1, inplace=True)
```
